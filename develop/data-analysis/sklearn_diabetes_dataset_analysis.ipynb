# Machine learning 

## Significato 
Dato un insieme di dati (dataset) usare questi ultimi per creare un algoritmo che sia capace in modo autonomo di estrarre  specifiche informazioni (features).

## Tipologie

* Supervised Learning: 
Dato un gruppo di dati, da noi scelti accuratamente, vogliamo predirre un informazione.
Vi sono due tipi di supervised learning: 
  * Classification: Dato un gruppo di dati vogliamo ottenere come risultato un gruppo di classi. (Discreto)Riconoscimento di caratteri a partire da immagini. Capire se un tumore é benigno o maligno a seconda della sua dimensione. Capire da dove viene ognuno di voi in base al vostro nome, cognome. Capire se fate sport o no in base a quanto siete muscolosi. 
  * Regression: La regressione ritorna un numero continuo. Qualé la lunghezza del salmone in base alla sua e peso. Individuare quante persone faranno sport nel prossimo mese tra di voi. Capire qualcosa sul futuro in base al presente.  

* Unsupervised Learning:
Dare ai dati un significato. Strutturare i dati. Se io voglio classificare cane e gatto e ho un daset di cani, gatti e orsi. Unsupervised learning ci aiuta a semplicare questo dataset in modo che non ci siano orsi perché non ci interessano. In generare capire cosa hanno in comune i dati e quali sono le informazioni piú importanti.
  * Clustering


## Citazione

Machine learning is about learning some properties of a data set and then testing those properties against another data set. A common practice in machine learning is to evaluate an algorithm by splitting a data set into two. We call one of those sets the training set, on which we learn some properties; we call the other set the testing set, on which we test the learned properties.

```
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes = datasets.load_diabetes()


# Use only one feature
diabetes_X = diabetes.data[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
