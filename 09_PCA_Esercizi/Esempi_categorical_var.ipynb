```
import pandas as pd
```

```
df = pd.read_csv('https://bit.ly/kaggletrain')
```

```
df.head()
```

```
df.shape
```

```
df.columns
```

```
df.isna().sum()
```

```
## Selecting features

# Survived = target
```

```
# not all rows!! df = df.loc[:.['Survived','Pclass','Sex','Embarked']]
```

```
df = df.loc[df.Embarked.notna(),['Survived','Pclass','Sex','Embarked']]
```

```
df.shape
```

```
df.head()
```

```
# 1 start by crossvalidating a survive, only with Pclass
# 2 How to use Pipeline
# 3 How to use One hot encoder and column transformer.
```

```
X = df.loc[:,['Pclass']]
y = df.Survived
```

```
X.shape
```

```
y.shape
```

```
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```
cross_val_score( LogisticRegression(), X,y, cv=5, scoring='accuracy').mean()
```

```
cross_val_score( LogisticRegression(), X,y, cv=10, scoring='accuracy').mean()
```

```
y.value_counts(normalize=True)
```

```
# I want to add more features to my model, and cross validate! Features are categorical!
```

```
#One hot encoding(= dummies encoding)
```

```
df.head()
```

```
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
```

```
ohe.fit_transform(df[['Embarked']])
```

```
ohe.categories_
```

```
X = df.drop('Survived',axis='columns')
```

```
X.head()
```

```
from sklearn.compose import make_column_transformer
```

```
column_trans = make_column_transformer(
              (OneHotEncoder(),['Sex','Embarked']),
              remainder = 'passthrough'    
)
```

```
column_trans.fit_transform(X)
```

```
from sklearn.pipeline import make_pipeline
```

```
pipe = make_pipeline(column_trans, LogisticRegression())
```

```
cross_val_score(pipe,X,y, cv=5, scoring='accuracy').mean()
```

```
X_new = X.sample(5, random_state=667)
X_new
```

```
pipe.fit(X,y)
```

```
pipe.predict(X_new)
```
